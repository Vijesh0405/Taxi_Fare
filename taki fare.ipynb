```python
mean_amount=df.describe()['amount']['mean']
mean_lo_pick=df.describe()['longitude_of_pickup']['mean']
mean_la_pick=df.describe()['latitude_of_pickup']['mean']
mean_lo_drop=df.describe()['longitude_of_dropoff']['mean']
mean_la_drop=df.describe()['latitude_of_dropoff']['mean']
```

```python
df['longitude_of_pickup']=df['longitude_of_pickup'].replace(0,df['longitude_of_pickup'].mean())
df['latitude_of_pickup']=df['latitude_of_pickup'].replace(0,df['latitude_of_pickup'].mean())
df['longitude_of_dropoff']=df['longitude_of_dropoff'].replace(0,df['longitude_of_dropoff'].mean())
df['latitude_of_dropoff']=df['latitude_of_dropoff'].replace(0,df['latitude_of_dropoff'].mean())
```

```python
df.head(20)
```

```python
df['diff_long'] = df['longitude_of_pickup']-df['longitude_of_dropoff']
df['diff_lati'] = df['latitude_of_pickup']-df['latitude_of_dropoff']
df['diff_long'] =df.diff_long.abs()
df['diff_lati'] =df.diff_lati.abs()
df.head(20)
```
